# Function Cretation for counting 

```python
def get_city_match_count(city):
    mask = data['city'] == city   # use the parameter
    return data[mask].shape[0]    # return only the row count
```

```python
get_city_match_count('Bangalore')
```

```python
get_city_match_count('Rajkot')
```
